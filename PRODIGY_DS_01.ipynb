```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('asthma_disease_data.csv')
```

```python
df.head()
```

# Histogram

```python
numeric_columns = df.select_dtypes(include='number').columns

df[numeric_columns].hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```

# Bar Chart

```python
avg_bmi_by_gender = df.groupby('Gender')['BMI'].mean()

avg_bmi_by_gender.plot(kind='bar', figsize=(8, 6))
plt.xlabel('Gender')
plt.ylabel('Average BMI')
plt.title('Average BMI by Gender')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation=0)  
plt.show()
```

```python
avg_bmi_by_diagnosis = df.groupby('Diagnosis')['BMI'].mean()
```

```python
avg_bmi_by_diagnosis.plot(kind='bar', figsize=(8, 6))
plt.xlabel('Diagnosis')
plt.ylabel('Average BMI')
plt.title('Average BMI by Diagnosis')
plt.xticks(ticks=avg_bmi_by_diagnosis.index, labels=['No Diagnosis', 'Diagnosis'], rotation=0)
plt.show()
```
